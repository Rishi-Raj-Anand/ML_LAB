```python
import numpy as np
import pandas as pd
import matplotlib.pyplot
```

```python
df = pd.read_csv('PlayTennis.csv')
print(df.head())
```

```python
def get_entropy(col):
    unique_values, counts = np.unique(col, return_counts=True)
    probabilities=counts/len(col)
    entropy=-np.sum(probabilities*np.log2(probabilities))
    return entropy
```

```python
def get_information_gain(colName,df,target):
    labels, counts = np.unique(df[colName], return_counts=True)
    probabilities=counts/len(df)
    entropy_list=[]
    for label in labels:
        temp_df=df[df[colName]==label]
        entropy_list.append(get_entropy(temp_df[target]))
    parent_entropy=get_entropy(df[target])
    ig=parent_entropy-np.sum(probabilities*entropy_list)
    return ig
```

```python
def decision_tree_root(df,target):
    columns=df.columns
    information_gain_data={}
    for col in columns:
        if(col!=target):
            information_gain_data[col]=get_information_gain(col,df,target)
    best_feature = max(information_gain_data, key=information_gain_data.get)

    return best_feature,information_gain_data[best_feature]
```

```python
root,ig=decision_tree_root(df,"play")

print(f"The Best column to split is '{root}' with information gain {ig}.")
```
